```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
pd.options.display.max_columns = None
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (8, 4)
```

```python
df = pd.read_csv('base/RISE/RR2017.txt', sep=';', encoding='ISO 8859-1')
```

```python
df.head()
```

```python
exploration = pd.DataFrame({
    'COLUMNS': df.columns,
    'TYPES': df.dtypes,
    'NA #': df.isna().sum(),
    'NA %': (df.isna().sum() / df.shape[0]) * 100
})
```

```python
exploration
```

```python
num_features = list(exploration[exploration['TYPES'] != 'object']['COLUMNS'])
cat_features = list(exploration[exploration['TYPES'] == 'object']['COLUMNS'])
```

```python
features_selected = [
    'Causa Afastamento 1',
    'Causa Afastamento 2',
    'Causa Afastamento 3',
    'Motivo Desligamento',
    'CBO Ocupação 2002',
    'CNAE 2.0 Classe',
    'Faixa Etária',
    'Faixa Hora Contrat',
    'Escolaridade após 2005',
    'Idade',
    'Município',
    'Nacionalidade',
    'Qtd Dias Afastamento',
    'Sexo Trabalhador',
    'Tipo Admissão',
    'Tipo Vínculo'
]
```

```python
df[features_selected].head()
```

```python
df[cat_features].head()
```

```python
df['Tipo Estab.1'] = df['Tipo Estab.1'].str.replace(' ', '')
df['Tipo Estab.1'] = df['Tipo Estab.1'].map({'CNPJ': 1, 'CEI': 0})
```

```python
df['Tempo Emprego'] = df['Tempo Emprego'].str.replace(',', '.')
df['Tempo Emprego'] = df['Tempo Emprego'].str.strip()

df['Tempo Emprego'] = df['Tempo Emprego'].astype('float32')
```

```python
cat_num_features = df[cat_features].dtypes[df[cat_features].dtypes == 'object'].index[5:]
```

```python
for col in cat_num_features:
    df[col] = df[col].str.replace(',', '.').astype(np.float64)
```

```python
df_num_features_preprocessed = df[df.dtypes[df.dtypes != 'object'].index]
```
